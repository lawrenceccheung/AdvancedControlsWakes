```python
def mean_vel_in_wake(self, radius, turb_loc = None, z = None, timespan = None, axis = 'x', component = 'u', verbose = True):
    """
    Calculates average velocity in the wake of a turbine

    Args in:
        hub_heigth (float): turbine hub height in m
        rot_diam (float): turbine rotor diameter in m
        turb_loc (list): x- and y- location of the turbine in m (default: [0, 0])
        dir (str): direction of the plane, either 'streamwise' (default) or 'slice' (i.e., cut-through of the flow)

    Args out:
        Utube (np.array): average wind speed in wake (dimensions: squeeze(num_time_steps, num_x_coor, num_cases) )
    """

    if z is None: z = list(self.z)
    if np.size(z) > 1:
        if axis == 'z': 
            vel = np.empty([self.get_time_index(timespan[-1], verbose = False)-self.get_time_index(timespan[0], verbose = False), 0])
            for zi in z:
                vel = np.append(vel, 
                                np.expand_dims(self.mean_vel_in_circle(turb_loc, radius, zi, timespan, component, verbose), axis=1), 
                                axis=1)
        else: 
            vel = np.empty((0, len(getattr(self, axis))))
            for zi in z:
                print([radius, turb_loc, zi, timespan, axis, component])
                vel = np.append(vel, self.vel_in_wake(radius, turb_loc, zi, timespan, axis, component, verbose), axis=0)
    else: 
        vel = self.vel_in_wake(radius, turb_loc, z, timespan, axis, component, verbose)
```

```python
def gauss_func_2d(X, v0, a, x0, y0, sigma, w = None, wy = None):
    """
    Simple or double gaussian function:
    Single:
        y = v0 - a*exp( -((x-x0) + (y-y0))^2 / (2*sigma^2) )
    Double:
        v0 - a*exp(-((x-x0-wx) + (y-y0-wy))^2 / (2*sigma^2) - a*exp( -((x-x0+wx) + (y-y0+wy))^2 / (2*sigma^2) )

    Args in:
        x (array-like): x vector
        y (array-like): y vector
        [v0, a, x0, sigma, w] (floats): Gauss inputs
            if w is not provided, single gaussian fit is applied

    Args out: 
        y (array): Gauss curve
    """
    
    x = X[:,0]
    y = X[:,1]
    
    if w is None:
        return v0 - a*np.exp(-(np.sqrt((x-x0)**2+(y-y0)**2))**2/(2*sigma**2))
    else: 
        return v0 - a*np.exp(-(np.sqrt((x-x0)**2+(y-y0)**2)-w)**2/(2*sigma**2)) - a*np.exp(-(np.sqrt((x-x0)**2+(y-y0)**2)+w)**2/(2*sigma**2))# - a*np.exp(-((x-x0+wx)**2 + (y-y0)**2)/(2*sigma**2))# - a*np.exp(-((x-x0)**2 + (y-y0+wy)**2)/(2*sigma**2)) - a*np.exp(-((x-x0)**2 + (y-y0+wy)**2)/(2*sigma**2))
    
    
def fit_to_2d_gauss(plane, z, x = None, y = None, p0 = None, fit = 'double', time = None, component = 'u', verbose = False):
    """
    Fit gaussian wake profile to cross section of wake.
    """
    
    if time is None: time = plane.time[-1]
    t_idx = plane.get_time_index(time, verbose=verbose)
    if x is None:
        x = plane.x
    if y is None:
        y = plane.y
    if np.size(z) == 1: 
        z = plane.get_plane(z, time, verbose=verbose)#get_plane_index(z, verbose=verbose)
        #z = plane.vel_planes[component][t_idx,:][z_idx*plane.x_N*plane.y_N:(z_idx+1)*plane.x_N*plane.y_N]
    elif len(np.shape(z)) > 1:
        z = np.reshape(z, -1)

    if p0 is None:
        v0 = np.max(plane.vel_planes['x'])
        a = np.max(plane.vel_planes['x']) - np.min(plane.vel_planes['x'])
        x0 = np.average(x)
        if plane.unit == 'm': sigma = 120
        else: sigma = 1
        if fit == 'double': p0 = [v0, a, x0, sigma/2, sigma/2]
        else: p0 = [v0, a, x0, sigma]

    if np.size(x)*np.size(y) == np.size(z):
        x, y = np.meshgrid(x, y)
        
    try: 
        popt, pcov = curve_fit(gauss_func_2d, np.vstack((x.ravel, y.ravel)), z.ravel(), p0)
    except:
        print(f'Warning! Optimal parameters not found at time {time}. Reverted to initial guess.')
        popt = p0

    return popt
    
```

```python
hub_height = 150
rot_diam = 240
time = 400
plane_idx = 0
wake_center_y = np.empty((ver_plane[0].z_N-1,len(ver_plane),len(ver_plane[0].time),))
for plane_idx in range(len(ver_plane)):
    for n in range(len(ver_plane[plane_idx].time)):
        for k in range(ver_plane[plane_idx].z_N-1):
            plane = ver_plane[plane_idx].get_plane(ver_plane[plane_idx].z[k+1],ver_plane[plane_idx].time[n])
            height = ver_plane[plane_idx].y[np.where(ver_plane[plane_idx].y<=2*hub_height)]
            plane_height = plane[np.where(height<=2*hub_height)]
            plane_avg = np.average(plane_height,axis=1)
            vel_sum = np.sum(plane_avg)
            wake_center_y[k,plane_idx,n] = np.interp(vel_sum/2,np.cumsum(plane_avg),height)

wake_center_x = np.empty((ver_plane[0].z_N-1,len(ver_plane),len(ver_plane[0].time),))
for plane_idx in range(len(ver_plane)):
    for n in range(len(ver_plane[plane_idx].time)):
        for k in range(ver_plane[plane_idx].z_N-1):
            plane = ver_plane[plane_idx].get_plane(ver_plane[plane_idx].z[k+1],ver_plane[plane_idx].time[n])
            plane_avg = np.average(plane,axis=0)
            vel_sum = np.sum(plane_avg)
            wake_center_x[k,plane_idx,n] = np.interp(vel_sum/2,np.cumsum(plane_avg),ver_plane[plane_idx].x)
```

```python
from scipy.optimize import curve_fit

def fit_sin(t,y):
    t = np.array(t)
    y = np.array(y)
    ff = np.fft.fftfreq(len(t), (t[1]-t[0]))
    Fyy = abs(np.fft.fft(y))
    guess_freq = abs(ff[np.argmax(Fyy[1:])+1])
    guess_amp = np.std(y) * 2.**0.5
    guess_offset = np.mean(y)
    guess = np.array([guess_amp, guess_freq, 0., guess_offset])

    def sin_fun(t,A,f,ph,offset):
        return A * np.sin(2*np.pi*f*t+ph) + offset
    
    popt, pcov = curve_fit(sin_fun,t,y,p0=guess)
    A,f,ph,offset = popt
    print(f'\
    Amplitude:    {A}\n\
    offset:       {offset}')
    curve = sin_fun(t,*popt)
    
    return popt, curve
for k in range(ver_plane[0].z_N-1):
    for n in range(len(ver_plane)-1):
        print(f'---   {case_folder[n+1]}   ---')
        popt,curve = fit_sin(ver_plane[n+1].time[299:-1],wake_center_x[k,n+1,300:len(ver_plane[n+1].time)])
```
